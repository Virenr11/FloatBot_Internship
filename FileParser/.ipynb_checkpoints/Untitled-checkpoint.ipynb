```python
doc = Document("C:\\Users\\viren\\OneDrive\\Desktop\\practical 5.docx")
chunks = []
for para in doc.paragraphs:
    chunks.append(para.text)
    for table in doc.tables:
            table_data = []
            for row in table.rows:
                row_data = []
                for cell in row.cells:
                    row_data.append(cell.text)
                table_data.append(row_data)
            chunks.append(table_data)
```

```python
doc = Document("C:\\Users\\viren\\OneDrive\\Desktop\\practical 5.docx")
chunk = []
for para in doc.paragraphs:
        if para.style.name.startswith('List'):
                chunk.append({'type': 'list', 'data': para.text})
        else:
                chunk.append(para.text)
        for table in doc.tables:
            table_data = []
            for row in table.rows:
                row_data = []
                for cell in row.cells:
                    row_data.append(cell.text)
                table_data.append(row_data)
            chunk.append(table_data)
        for img in doc.inline_shapes:
            chunk.append({'type': 'image', 'data': img._inline.docx.blob})
```
